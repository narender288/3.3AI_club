```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('gym_exercise_weight_loss.csv')
```

```python
data
```

```python
#Task1 - manipulate the dataset and extract the Hours exercised and Weight Lost
E = data['Hours Exercised (E)']
W = data['Weight Loss (W)']
X = np.array(E)
Y = np.array(W)
X = X.reshape(-1,1)
#converting data types and structure to fir our model to be used next
```

```python
#Let's plot the data to get an idea of how it behaves I'm I'm
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='red', label='Data')
plt.xlabel('E')
plt.ylabel('W')
plt.title('Scatter Plot of E vs W')
plt.legend()
plt.show()
```

Now that we have our data ready, time to use a regression model to find the best fit line to our dataset as it seems that astraight line might fit the data properly

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Task 2 - Create an instance of LinearRegression and fit X and Y to it
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Task 3 - Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Task 4 - Calculate the Mean Squared Error
mse = mean_squared_error(Y, Y_pred)

# Output the results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Mean Squared Error: {mse}")
```

```python
#Let's see how our function fits the model graphically
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='red', label='Data')
plt.plot(X, Y_pred, color='blue', label='Linear Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Linear Regression\nMean Squared Error: {mse:.4f}')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# Define the degree of the polynomial (manually change this value)
degree = 2  # Change this to the desired degree of the polynomial (1-10)

# Create polynomial features and fit model
polynomial_features = PolynomialFeatures(degree=degree)
linear_regression = LinearRegression()
model = make_pipeline(polynomial_features, linear_regression)

# Fit the model
model.fit(X, Y)

# Generate predictions
Y_pred = model.predict(X)

# Calculate Mean Squared Error (loss)
mse = mean_squared_error(Y, Y_pred)

# Generate a smooth curve for the plot
X_range = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
Y_range_pred = model.predict(X_range)

# Plot the original data and the polynomial fit
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='red', label='Data')
plt.plot(X_range, Y_range_pred, color='blue', label=f'Polynomial degree {degree}')
plt.xlabel('Hours Exercised')
plt.ylabel('Weight Lost')
plt.title(f'Polynomial Regression (degree={degree})\nMean Squared Error: {mse:.4f}')
plt.legend()
plt.show()
```
